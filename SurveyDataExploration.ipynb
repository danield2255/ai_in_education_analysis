# AI In Education Exploration

Since AI tools have become widely popularized in the early 2020's, one area that their usage has become popular in is education. We will look at a small survey of data collected from 2nd and 3rd year students in the Faculty of Cybernetics, Statistics and Economic Informatics in Romania. There are only 91 survey responses so the data is decently small scale, but the insights are still valuable in understanding what emerging technical workers think of Artificial Intelligence. 

------------

```python
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt

df = pl.read_csv("Survey_AI.csv")
```

```python
print("Shape:", df.shape)
print("Columns:", df.columns)
```

## Preprocessing

Here we will adjust our column names to be more readable, and we will conduct general cleaning.

```python
df_new = df.rename({"Q1.AI_knowledge": "ai_knowledge_1-10", 
                "Q2.AI_sources": "ai_learning_sources", 
                "Q2#1.Internet": "learn_from_internet", 
                "Q2#2.Books/Papers": "learn_from_books_papers",
                "Q2#3.Social_media": "learn_from_social_media",
                "Q2#4.Discussions": "learn_from_discussions",
                "Q2#5.NotInformed": "not_informed_on_ai", 
                "Q3#1.AI_dehumanization": "agree_ai_dehumanization", 
                "Q3#2.Job_replacement": "agree_job_replacement",
                "Q3#3.Problem_solving": "agree_ai_problem_solving",
                "Q3#4.AI_rulling_society": "agree_ai_ruling_society",
                "Q4#1.AI_costly": "agree_ai_costly",
                "Q4#2.Economic_crisis": "agree_ai_economic_crisis",
                "Q4#3.Economic_growth": "agree_ai_economic_growth",
                "Q4#4.Job_loss": "agree_ai_job_loss",
                "Q5.Feelings": "feelings_about_ai",
                "Q6.Domains": "domains_of_ai_impact",
                "Q6#1.Education": "ai_impact_education",
                "Q6#2.Medicine": "ai_impact_medicine",
                "Q6#3.Agriculture": "ai_impact_agriculture",
                "Q6#4.Constructions": "ai_impact_construction",
                "Q6#5.Marketing": "ai_impact_marketing",
                "Q6#6.Administration": "ai_impact_administration",
                "Q6#7.Art": "ai_impact_art",
                "Q7.Utility_grade" : "grade_ai_in_education_1-10",
                "Q8.Advantage_teaching": "ai_main_advantage_teaching",
                "Q9.Advantage_learning": "ai_main_advantage_learning",
                "Q10.Advantage_evaluation ": "ai_main_advantage_evaluation",
                "Q11.Disadvantage_educational_process": "ai_main_disadvantage_educational_process",
                "Q12.Gender": "gender",
                "Q13.Year_of_study": "year_of_study",
                "Q14.Major": "major",
                "Q15.Passed_exams": "passed_exams",
                "Q16.GPA": "gpa"})
```

We will replace some encoded values with better labels here. 

```python

df_new = df_new.with_columns(
    pl.when(pl.col("gender") == 1).then(pl.lit("Female"))
      .when(pl.col("gender") == 2).then(pl.lit("Male"))
      .otherwise(pl.col("gender"))
      .alias("gender")
)
```

```python
df_new = df_new.with_columns(
    pl.when(pl.col("major") == 1).then(pl.lit("Economic Cybernetics"))
      .when(pl.col("major") == 2).then(pl.lit("Statistics and Economic Forecasting"))
      .when(pl.col("major") == 3).then(pl.lit("Economic Informatics"))
      .otherwise(pl.col("major"))
      .alias("major")
)
```

```python
## Demographic Analysis: Student Perspectives on AI

# 1. AI Knowledge by Demographics
print("=== AI Knowledge by Gender ===")
print(df_new.group_by("gender").agg(pl.col("ai_knowledge_1-10").mean()))

print("\n=== AI Knowledge by Major ===")
print(df_new.group_by("major").agg(pl.col("ai_knowledge_1-10").mean()))

print("\n=== AI Knowledge by Year of Study ===")
print(df_new.group_by("year_of_study").agg(pl.col("ai_knowledge_1-10").mean()))

# 2. Visualize AI Knowledge across Demographics
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# By Gender
df_new.to_pandas().boxplot(column="ai_knowledge_1-10", by="gender", ax=axes[0])
axes[0].set_title("AI Knowledge by Gender")
axes[0].set_xlabel("Gender")
axes[0].set_ylabel("Knowledge (1-10)")

# By Major
df_new.to_pandas().boxplot(column="ai_knowledge_1-10", by="major", ax=axes[1])
axes[1].set_title("AI Knowledge by Major")
axes[1].set_xlabel("Major")
axes[1].set_ylabel("Knowledge (1-10)")

# By Year of Study
df_new.to_pandas().boxplot(column="ai_knowledge_1-10", by="year_of_study", ax=axes[2])
axes[2].set_title("AI Knowledge by Year of Study")
axes[2].set_xlabel("Year")
axes[2].set_ylabel("Knowledge (1-10)")

plt.tight_layout()
plt.show()

# 3. AI Usefulness in Education by Demographics
print("\n=== Perceived AI Usefulness in Education ===")
print(df_new.group_by("gender").agg(pl.col("grade_ai_in_education_1-10").mean()))
print(df_new.group_by("major").agg(pl.col("grade_ai_in_education_1-10").mean()))
```
