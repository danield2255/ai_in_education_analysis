## EDA

Let us ask some targetted questions of the data. We will start with the following questions and evaluate them from a number of perspectives:
1. Generally how informed do students seem to be about AI?
2. Do students of different academic performance seem to think of AI differently?
3. What are students most concerned about when it pertains to AI?

### 1. Generally how informed do students seem to be about AI?
We will start with looking at the distribution of responses to question. 

```python
plt.hist(df_new.select("ai_knowledge_1-10"), bins=10, edgecolor='black')
plt.title("Distribution of AI Knowledge (1-10)")
plt.xlabel("AI Knowledge (1-10)")
plt.ylabel("Frequency")
plt.show()
```

Though the sample is not very large, we can see that the ranked knowledge of AI among respondents follows a pretty normal distribution. We may want to create a sub-group of respondents with a knowledge ranked 7 or higher, and consider these people the more well informed. Then perhaps we can see where those most informed people learn from so as to advise less informed students where they can learn more about AI. 

```python
df_new = df_new.with_columns(
    pl.when(pl.col("ai_knowledge_1-10") >= 7)
    .then(pl.lit("Well informed"))
    .otherwise(pl.lit("Less informed"))
    .alias("ai_informed"))

informed = df_new.filter(pl.col("ai_informed") == "Well informed")
```

```python
learning_sources = {"Internet": informed["learn_from_internet"].sum(),
    "Books/Papers": informed["learn_from_books_papers"].sum(),
    "Social Media": informed["learn_from_social_media"].sum(),
    "Discussions": informed["learn_from_discussions"].sum()}

sources = list(learning_sources.keys())
counts = list(learning_sources.values())

plt.bar(sources, counts, edgecolor='black', color='steelblue')
plt.title("AI Learning Sources Among Well Informed Students")
plt.xlabel("Learning Source")
plt.ylabel("Number of Students")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("Learning Sources Summary (Well Informed Students):")
for source, count in learning_sources.items():
    print(f"{source}: {count}")
```

Here the data is general but we see that most well informed students are learning about AI through the internet and Social Media accounts. These might be good places to start learning (however it should be noted that perhaps more reliable and academicly consistent information will come from published books and academic papers!)

### 2. Do students of different academic performance seem to think of AI differently?
Here will will look at students by GPA and investigate the sentiment toward AI of students with different academic performance. We will transform the GPA buckets to be a little smaller as to consolidate some groups.

```python
df_new = df_new.with_columns(
    pl.when(pl.col("gpa").is_between(5, 6)).then(pl.lit("5-6"))
    .when(pl.col("gpa").is_between(6, 7)).then(pl.lit("6-7"))
    .when(pl.col("gpa").is_between(7, 8)).then(pl.lit("7-8"))
    .when(pl.col("gpa").is_between(8, 9)).then(pl.lit("8-9"))
    .when(pl.col("gpa").is_between(9, 10)).then(pl.lit("9-10"))
    .alias("gpa_buckets"))


df_new.group_by("gpa_buckets").agg([pl.len().alias("num_students"), 
                                    pl.col("ai_knowledge_1-10").mean().alias("avg_ai_knowledge"), 
                                    pl.col("not_informed_on_ai").sum().alias("num_not_informed"),
                                    (pl.col("not_informed_on_ai").sum()/pl.len()).alias("proportion_not_informed"),
                                    pl.col("agree_ai_dehumanization").mean().alias("mean_level_agree_ai_dehumanization"),
                                    pl.col("agree_job_replacement").mean().alias("mean_level_agree_job_replacement"),
                                    pl.col("agree_ai_problem_solving").mean().alias("mean_level_agree_ai_problem_solving"),
                                    pl.col("agree_ai_ruling_society").mean().alias("mean_level_agree_ai_ruling_society"),
                                    pl.col("agree_ai_costly").mean().alias("mean_level_agree_ai_costly"),
                                    pl.col("agree_ai_economic_crisis").mean().alias("mean_level_agree_ai_economic_crisis"),
                                    pl.col("agree_ai_economic_growth").mean().alias("mean_level_agree_ai_economic_growth"),
                                    pl.col("agree_ai_job_loss").mean().alias("mean_level_agree_ai_job_loss"),
                                    pl.col("feelings_about_ai").mode().alias("feelings_about_ai_distribution")
                                    ]).sort("gpa_buckets")
```

We can see from the output there are few interesting things to call out:
- We see there is mostly a positive association between gpa and average AI Knowledge. 
- There are only 4 students in the lowest GPA group, but those students seemed to have the most negative general view of AI on the whole. They also mostly seem to be fearful of AI
- The students in the highest GPA group seemed to have the most positive or optimistic view of AI on the whole. 
- Aside from the few students in the lowest GPA buckets most students seem primarily curious
